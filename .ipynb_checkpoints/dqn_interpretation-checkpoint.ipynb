# DQN Policy Interpretation via Schelling Benchmark

This notebook performs two complementary interpretability analyses on the two DQN agents (Type A and Type B) trained in the Schelling-like segregation model from:

> *Segregation dynamics with reinforcement learning and agent-based modeling*

## Analyses

1. **Behavioral comparison vs. Schelling rule** — For every possible neighborhood composition (fraction of same-type neighbors $f_s$), we compare the DQN agent's most frequent action to the classical Schelling rule (stay if $f_s \geq \theta$, move otherwise). This reveals whether the learned policy converges to, deviates from, or refines the analytic rule.

2. **Q-value sweep as a function of neighborhood composition** — We systematically vary the local patch seen by each agent (fraction of same-type neighbors $f_s \in [0,1]$, fraction of opposite-type $f_o \in [0,1-f_s]$) and record the raw Q-values for each action. This produces a 2-D or 1-D landscape of the value function, showing exactly how the network weights each local social context.

---

**Prerequisites:** run the training notebook (or `schelling_example.py`) first so that experiment weights (`crystal.npy.gz` or the `Schelling` object) are available in memory.

## 0. Imports & setup

```python
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)

import os, random, shutil
from multiprocessing import Lock
import numpy as np
import torch
import torch.nn.functional as F

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.ticker as mticker
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
from scipy.ndimage import label as scipy_label

# ── Original classes (unchanged) ─────────────────────────────────────────────
from environment import Environment
from schelling_example import Schelling, play

# ── Reproducibility ───────────────────────────────────────────────────────────
SEED = 1
np.random.seed(SEED)
random.seed(SEED)
torch.manual_seed(SEED)

matplotlib.rcParams['figure.dpi'] = 120
matplotlib.rcParams['axes.spines.top']   = False
matplotlib.rcParams['axes.spines.right'] = False

# ── Action labels (paper convention: 0=up 1=left 2=down 3=right 4=stay) ──────
ACTION_NAMES  = ['Up', 'Left', 'Down', 'Right', 'Stay']
ACTION_DIRS   = [(-1,0),(0,-1),(1,0),(0,1),(0,0)]
N_ACTIONS     = 5

# ── Color palette ─────────────────────────────────────────────────────────────
COL_A    = '#4a90d9'
COL_B    = '#e05252'
COL_STAY = '#2ca02c'
COL_MOVE = '#d62728'

print('All imports OK.')
print(f'PyTorch {torch.__version__} | CUDA: {torch.cuda.is_available()}')
```

## 1. Recreate (or reload) trained society

We rebuild the `Schelling` environment with the **same parameters** used in training and run a short warm-up so the networks are trained. If you already have a trained `society` object from the previous notebook you can skip the `run_instrumented` call and just pass the object directly.

```python
# ── Paper-default hyper-parameters ───────────────────────────────────────────
GRID_SIZE   = (50, 50)
P_HUNTER    = 0.05
P_PREY      = 0.0
AGENT_RANGE = 5         # FOV radius → (2R+1)² = 121 input dims
MAX_AGE     = 100
ALPHA       = 1.0
BETA        = 1.0
GAMMA_R     = 1.0       # renamed to avoid clash with Python's γ
ITERATIONS  = 250

EXP_NAME = 'interp_exp'
if os.path.isdir(EXP_NAME):
    shutil.rmtree(EXP_NAME)

np.random.seed(SEED); random.seed(SEED); torch.manual_seed(SEED)

society = Schelling(
    GRID_SIZE,
    p_hunter      = P_HUNTER,
    p_prey        = P_PREY,
    prey_reward   = 0,
    agent_max_age = MAX_AGE,
    agent_range   = AGENT_RANGE,
    max_iteration = ITERATIONS,
    lock          = Lock(),
    name          = EXP_NAME,
    alpha         = ALPHA,
    beta          = BETA,
    gamma         = GAMMA_R,
)

print('Society created — running training loop...')
play(society)   # uses the authors' unmodified play() — trains the networks
print('Training done.')
```

## 2. Extract trained networks & helper utilities

We pull the two `DQN` (or equivalent) networks out of the `Schelling` object. The exact attribute name depends on the authors' implementation; we probe common names.

```python
def _get_mind(society, agent_type_val):
    """
    Retrieve the policy network for agent_type_val (-1→A, +1→B).
    Tries several attribute naming conventions used in the paper's code.
    """
    candidates_A = ['A_mind', 'mind_A', 'hunter_mind', 'policyA', 'netA', 'dqn_A']
    candidates_B = ['B_mind', 'mind_B', 'prey_mind',   'policyB', 'netB', 'dqn_B']
    candidates   = candidates_A if agent_type_val == -1 else candidates_B
    for attr in candidates:
        net = getattr(society, attr, None)
        if net is not None:
            return net
    # Fallback: inspect all attributes for torch.nn.Module
    import torch.nn as nn
    modules = [(k, v) for k, v in vars(society).items() if isinstance(v, nn.Module)]
    if len(modules) >= 2:
        idx = 0 if agent_type_val == -1 else 1
        print(f'  Auto-detected network: {modules[idx][0]}')
        return modules[idx][1]
    raise AttributeError(
        'Could not find the DQN network in the Schelling object.\n'
        f'Available attributes: {[k for k in vars(society).keys()]}'
    )

net_A = _get_mind(society, -1)
net_B = _get_mind(society,  1)

net_A.eval()
net_B.eval()

print('Networks extracted.')
print(f'  Type A network: {type(net_A).__name__}')
print(f'  Type B network: {type(net_B).__name__}')
print(f'  Parameters A  : {sum(p.numel() for p in net_A.parameters()):,}')
print(f'  Parameters B  : {sum(p.numel() for p in net_B.parameters()):,}')
```

## 3. Canonical patch builder

The DQN input is a **$(2R+1)\times(2R+1)$ neighbourhood patch** flattened to a 1-D vector (or passed as a 2-D map into a CNN). We build *synthetic* patches parameterised by:

| Symbol | Meaning |
|--------|--------|
| $f_s$ | fraction of cells occupied by **same-type** agents |
| $f_o$ | fraction of cells occupied by **opposite-type** agents |
| $f_e = 1 - f_s - f_o$ | fraction of **empty** cells |

The agent itself is always placed at the centre. Cells are distributed randomly across the remaining $(2R+1)^2 - 1$ positions.

```python
R         = AGENT_RANGE          # 5  → 11×11 patch
PATCH_W   = 2 * R + 1            # 11
N_CELLS   = PATCH_W ** 2         # 121
N_NEIGH   = N_CELLS - 1          # 120  (excludes centre)
CX        = R                    # centre row
CY        = R                    # centre col

# ── Agent-type encoding: A=-1, empty=0, B=+1, prey=2 ─────────────────────────
SAME_VAL  = {'A': -1, 'B':  1}
OPP_VAL   = {'A':  1, 'B': -1}


def make_patch(agent_type: str, f_same: float, f_opp: float,
               rng: np.random.Generator = None, n_avg: int = 1) -> np.ndarray:
    """
    Build a (PATCH_W, PATCH_W) synthetic neighbourhood patch.

    Parameters
    ----------
    agent_type : 'A' or 'B'
    f_same     : fraction of neighbour cells filled by same-type  [0,1]
    f_opp      : fraction of neighbour cells filled by opposite   [0,1-f_same]
    rng        : optional numpy Generator for reproducibility
    n_avg      : number of random patches to average before network query
                 (reduces variance from spatial arrangement)

    Returns
    -------
    patch : (PATCH_W, PATCH_W) float array
    """
    if rng is None:
        rng = np.random.default_rng(SEED)

    sv = SAME_VAL[agent_type]
    ov = OPP_VAL[agent_type]

    n_same = max(0, min(N_NEIGH, round(f_same * N_NEIGH)))
    n_opp  = max(0, min(N_NEIGH - n_same, round(f_opp  * N_NEIGH)))

    if n_avg == 1:
        patch = np.zeros((PATCH_W, PATCH_W), dtype=np.float32)
        patch[CX, CY] = sv                    # self at centre
        offsets = [(r, c) for r in range(PATCH_W) for c in range(PATCH_W)
                   if not (r == CX and c == CY)]
        chosen  = rng.choice(len(offsets), n_same + n_opp, replace=False)
        for k, idx in enumerate(chosen):
            r, c = offsets[idx]
            patch[r, c] = sv if k < n_same else ov
        return patch
    else:
        return np.stack([make_patch(agent_type, f_same, f_opp, rng, 1)
                         for _ in range(n_avg)], axis=0).mean(axis=0).astype(np.float32)


def patch_to_tensor(patch: np.ndarray) -> torch.Tensor:
    """
    Convert a (H, W) patch to the tensor format expected by the DQN.
    The original code flattens the patch: shape (1, N_CELLS).
    If the network has a Conv layer we keep it 2-D: (1, 1, H, W).
    We detect automatically.
    """
    flat = torch.tensor(patch.flatten(), dtype=torch.float32).unsqueeze(0)  # (1, 121)
    return flat


def get_qvals(net: torch.nn.Module, patch: np.ndarray) -> np.ndarray:
    """
    Forward pass through `net` and return Q-values as a (N_ACTIONS,) numpy array.
    Handles both flat-input MLP and conv-input CNN transparently.
    """
    t = patch_to_tensor(patch)                 # (1, 121)
    with torch.no_grad():
        try:
            q = net(t)
        except RuntimeError:
            # CNN path: try (1, 1, H, W)
            t2 = torch.tensor(patch, dtype=torch.float32).unsqueeze(0).unsqueeze(0)
            q  = net(t2)
    return q.squeeze().numpy()


def greedy_action(net: torch.nn.Module, patch: np.ndarray) -> int:
    """Return argmax Q action index."""
    return int(np.argmax(get_qvals(net, patch)))


# ── Quick sanity check ────────────────────────────────────────────────────────
test_patch = make_patch('A', f_same=0.5, f_opp=0.1)
q_test     = get_qvals(net_A, test_patch)
print(f'Sanity check — Q-values for type A, f_s=0.5, f_o=0.1:')
for a, q in zip(ACTION_NAMES, q_test):
    print(f'  {a:6s}: {q:+.4f}')
print(f'  => greedy action: {ACTION_NAMES[int(np.argmax(q_test))]}')
```

## 4. Schelling rule baseline

The classical Schelling rule is a **threshold rule**: an agent stays if the fraction of same-type neighbours exceeds a tolerance threshold $\theta$, and moves (to a random empty cell) otherwise.

$$\pi_{\text{Schelling}}(f_s) = \begin{cases} \text{stay}  & f_s \geq \theta \\ \text{move} & f_s < \theta \end{cases}$$

We compare this baseline to the DQN policy at the **action level** (stay vs. move) and at the **Q-value level**.

```python
SCHELLING_THETA = 0.333   # standard 1/3 threshold used in the paper

def schelling_action(f_same: float, theta: float = SCHELLING_THETA) -> int:
    """
    Returns ACTION index: 4 (stay) if f_same >= theta, else 0 (move = Up as proxy).
    In the binary Schelling model the direction of movement is random;
    we encode 'move' as the *probability of NOT choosing Stay*.
    """
    return 4 if f_same >= theta else 0   # 4=Stay in paper convention


def is_stay(action_idx: int) -> bool:
    return action_idx == 4

print(f'Schelling threshold θ = {SCHELLING_THETA}')
```

# Analysis 1 — Behavioral comparison vs. Schelling rule

### Protocol

1. Sweep $f_s \in [0,1]$ in steps of $0.05$ (21 points) with $f_o = 0$ (only same-type and empty neighbours — pure homophily dimension).
2. For each $f_s$, build $N=50$ random patches and query the DQN. Record the **fraction of times the DQN chooses Stay**.
3. Overlay the hard Schelling threshold at $\theta$.
4. Compute **agreement rate** between DQN and Schelling rule.

```python
N_SAMPLES  = 50           # patches per (f_s, f_o) point
FS_GRID    = np.linspace(0, 1, 21)   # 0, 0.05, …, 1.0
rng_sweep  = np.random.default_rng(42)

def stay_prob_1d(net, agent_type, f_opp=0.0, n_samples=N_SAMPLES):
    """
    For each f_s value, compute P(DQN picks Stay) averaged over n_samples patches.
    f_opp is fixed (typically 0 for the 1-D analysis).
    """
    probs = []
    for fs in FS_GRID:
        # clip to valid range
        fo = min(f_opp, max(0, 1 - fs - 1/N_NEIGH))
        stays = 0
        for _ in range(n_samples):
            p  = make_patch(agent_type, fs, fo, rng=rng_sweep)
            a  = greedy_action(net, p)
            stays += int(is_stay(a))
        probs.append(stays / n_samples)
    return np.array(probs)

print('Computing DQN stay-probabilities (1-D sweep over f_s)...')
stay_A_1d = stay_prob_1d(net_A, 'A')
stay_B_1d = stay_prob_1d(net_B, 'B')

# Schelling reference
schelling_stay = np.array([1.0 if fs >= SCHELLING_THETA else 0.0 for fs in FS_GRID])

# Agreement rate
agree_A = float(np.mean((stay_A_1d >= 0.5) == (schelling_stay == 1.0)))
agree_B = float(np.mean((stay_B_1d >= 0.5) == (schelling_stay == 1.0)))

print(f'Agreement with Schelling rule (θ={SCHELLING_THETA}):  A={agree_A:.1%}  B={agree_B:.1%}')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4.5), sharey=True)

for ax, stay_dqn, col, label, agree in [
    (axes[0], stay_A_1d, COL_A, 'Type A', agree_A),
    (axes[1], stay_B_1d, COL_B, 'Type B', agree_B),
]:
    # Schelling step function
    ax.step(FS_GRID, schelling_stay, where='post',
            color='black', lw=2.5, ls='--', label=f'Schelling rule (θ={SCHELLING_THETA})',
            zorder=3)
    ax.axvline(SCHELLING_THETA, color='black', lw=0.8, alpha=0.4)

    # DQN probability curve
    ax.plot(FS_GRID, stay_dqn, color=col, lw=2.5, marker='o', ms=6,
            label=f'DQN {label} (P(Stay))', zorder=4)

    # Shade agreement / disagreement regions
    ax.fill_between(FS_GRID, stay_dqn, schelling_stay,
                    where=(stay_dqn >= 0.5) != (schelling_stay == 1.0),
                    alpha=0.18, color='red',  label='Disagreement region')
    ax.fill_between(FS_GRID, stay_dqn, schelling_stay,
                    where=(stay_dqn >= 0.5) == (schelling_stay == 1.0),
                    alpha=0.10, color='green', label='Agreement region')

    ax.set_xlabel('Fraction of same-type neighbours $f_s$', fontsize=11)
    ax.set_ylabel('P(Stay)', fontsize=11)
    ax.set_title(f'{label} vs. Schelling rule\n(agreement = {agree:.1%})', fontsize=12)
    ax.set_xlim(0, 1); ax.set_ylim(-0.05, 1.15)
    ax.legend(fontsize=8, loc='upper left')
    ax.grid(alpha=0.3)
    ax.text(SCHELLING_THETA + 0.01, 0.05, f'θ={SCHELLING_THETA}', fontsize=8, color='gray')

plt.suptitle('DQN Policy vs. Classical Schelling Rule  (f_o = 0)', fontsize=14, y=1.01)
plt.tight_layout()
plt.savefig('behavioral_comparison_1d.pdf', bbox_inches='tight')
plt.show()
print('Saved behavioral_comparison_1d.pdf')
```

### 4.2  2-D behavioral map — varying both $f_s$ and $f_o$

The Schelling model ignores the opposite-type fraction; the DQN may not. We now sweep both $f_s$ and $f_o$ jointly and plot the **DQN's P(Stay)** as a 2-D heatmap, with the Schelling decision boundary overlaid.

```python
FS_2D = np.linspace(0, 1, 13)    # coarser grid for speed
FO_2D = np.linspace(0, 1, 13)
N_AVG = 30

def stay_prob_2d(net, agent_type, n_samples=N_AVG):
    """Return (len(FO_2D), len(FS_2D)) array of P(Stay)."""
    grid = np.full((len(FO_2D), len(FS_2D)), np.nan)
    rng  = np.random.default_rng(7)
    for j, fs in enumerate(FS_2D):
        for i, fo in enumerate(FO_2D):
            if fs + fo > 1.0 + 1e-6:
                continue   # infeasible combination
            stays = 0
            for _ in range(n_samples):
                p = make_patch(agent_type, fs, fo, rng=rng)
                stays += int(is_stay(greedy_action(net, p)))
            grid[i, j] = stays / n_samples
    return grid

print('Computing 2-D behavioral maps (this may take ~30 s)...')
stay2d_A = stay_prob_2d(net_A, 'A')
stay2d_B = stay_prob_2d(net_B, 'B')
print('Done.')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

for ax, grid2d, col, label in [
    (axes[0], stay2d_A, COL_A, 'Type A'),
    (axes[1], stay2d_B, COL_B, 'Type B'),
]:
    # Mask infeasible region
    masked = np.ma.masked_invalid(grid2d)

    im = ax.imshow(
        masked, origin='lower', aspect='auto',
        extent=[FS_2D[0], FS_2D[-1], FO_2D[0], FO_2D[-1]],
        cmap='RdYlGn', vmin=0, vmax=1, interpolation='bilinear'
    )
    plt.colorbar(im, ax=ax, label='P(Stay)', shrink=0.85)

    # Schelling boundary: vertical line at f_s = theta (independent of f_o)
    ax.axvline(SCHELLING_THETA, color='black', lw=2, ls='--',
               label=f'Schelling θ={SCHELLING_THETA}')

    # Infeasible boundary f_s + f_o = 1
    fs_bound = np.linspace(0, 1, 100)
    ax.plot(fs_bound, 1 - fs_bound, 'k:', lw=1.2, label='$f_s+f_o=1$ boundary')
    ax.fill_between(fs_bound, 1 - fs_bound, 1.0, alpha=0.15, color='gray')

    ax.set_xlabel('Same-type fraction $f_s$', fontsize=11)
    ax.set_ylabel('Opposite-type fraction $f_o$', fontsize=11)
    ax.set_title(f'{label}: P(Stay) over $(f_s, f_o)$ space', fontsize=12)
    ax.legend(fontsize=8, loc='upper right')
    ax.set_xlim(0, 1); ax.set_ylim(0, 1)

plt.suptitle('2-D Behavioral Map: DQN vs. Schelling boundary', fontsize=14, y=1.01)
plt.tight_layout()
plt.savefig('behavioral_comparison_2d.pdf', bbox_inches='tight')
plt.show()
print('Saved behavioral_comparison_2d.pdf')
```

# Analysis 2 — Q-value sweep as a function of neighbourhood composition

Instead of the binary *stay/move* decision, here we inspect the **raw Q-values** for every action as a function of $f_s$. This reveals:

- The **value landscape**: how much future reward each action is expected to yield at each social context.
- Whether the DQN develops a **soft threshold** (smooth Q crossing) or a **hard threshold** (discontinuous argmax switch) — analogous to the Schelling step function.
- The **advantage** of Stay over the best movement action as a function of $f_s$.

### 5.1  Q-value curves along the $f_s$ axis  ($f_o = 0$)

```python
FS_FINE  = np.linspace(0, 1, 41)   # 41 points
N_Q      = 30                       # patches per point

def qval_sweep_1d(net, agent_type, f_opp=0.0, n_samples=N_Q):
    """
    Returns array of shape (len(FS_FINE), N_ACTIONS) with mean Q-values.
    """
    rng = np.random.default_rng(99)
    out = np.zeros((len(FS_FINE), N_ACTIONS))
    for k, fs in enumerate(FS_FINE):
        fo = min(f_opp, max(0.0, 1 - fs - 1/N_NEIGH))
        qs = np.zeros(N_ACTIONS)
        for _ in range(n_samples):
            p   = make_patch(agent_type, fs, fo, rng=rng)
            qs += get_qvals(net, p)
        out[k] = qs / n_samples
    return out

print('Computing Q-value sweeps (1-D, f_o=0)...')
qA_1d = qval_sweep_1d(net_A, 'A')
qB_1d = qval_sweep_1d(net_B, 'B')
print('Done.')
```

```python
ACTION_COLORS = ['#1f77b4','#ff7f0e','#2ca02c','#9467bd','#d62728']

fig, axes = plt.subplots(2, 2, figsize=(14, 9))

for row_idx, (qvals, col, label) in enumerate([
    (qA_1d, COL_A, 'Type A'),
    (qB_1d, COL_B, 'Type B'),
]):

    # ── Left: all 5 Q-value curves ───────────────────────────────────────────
    ax = axes[row_idx, 0]
    for a in range(N_ACTIONS):
        lw   = 3 if a == 4 else 1.5    # thicker for Stay
        ls   = '-' if a == 4 else '--'
        ax.plot(FS_FINE, qvals[:, a],
                color=ACTION_COLORS[a], lw=lw, ls=ls,
                label=f'Q({ACTION_NAMES[a]})')

    ax.axvline(SCHELLING_THETA, color='black', lw=1.5, ls=':', alpha=0.7,
               label=f'Schelling θ={SCHELLING_THETA}')

    # Mark Q-crossing point: where Q(Stay) becomes the max
    greedy = np.argmax(qvals, axis=1)
    switch = np.where(np.diff((greedy == 4).astype(int)))[0]
    for sw in switch:
        ax.axvline(FS_FINE[sw], color='gray', lw=1, ls='-.',
                   label=f'DQN switch at f_s={FS_FINE[sw]:.2f}')

    ax.set_xlabel('$f_s$ (same-type fraction)', fontsize=10)
    ax.set_ylabel('Q-value', fontsize=10)
    ax.set_title(f'{label} — Q-values for each action', fontsize=11)
    ax.legend(fontsize=7.5, ncol=2)
    ax.grid(alpha=0.25)
    ax.set_xlim(0, 1)

    # ── Right: advantage of Stay over best movement action ───────────────────
    ax = axes[row_idx, 1]
    best_move_q = np.max(qvals[:, :4], axis=1)      # best of Up/Left/Down/Right
    stay_q      = qvals[:, 4]
    advantage   = stay_q - best_move_q

    ax.fill_between(FS_FINE, 0, advantage,
                    where=advantage >= 0, alpha=0.3, color=COL_STAY, label='Stay preferred')
    ax.fill_between(FS_FINE, 0, advantage,
                    where=advantage < 0,  alpha=0.3, color=COL_MOVE, label='Move preferred')
    ax.plot(FS_FINE, advantage, color=col, lw=2.5)
    ax.axhline(0, color='black', lw=1.2)
    ax.axvline(SCHELLING_THETA, color='black', lw=1.5, ls=':', alpha=0.7,
               label=f'Schelling θ={SCHELLING_THETA}')

    ax.set_xlabel('$f_s$ (same-type fraction)', fontsize=10)
    ax.set_ylabel('Q(Stay) − Q(best move)', fontsize=10)
    ax.set_title(f'{label} — Stay advantage over best movement', fontsize=11)
    ax.legend(fontsize=8)
    ax.grid(alpha=0.25)
    ax.set_xlim(0, 1)

plt.suptitle('Q-value Landscape vs. Neighbourhood Composition  ($f_o=0$)', fontsize=14, y=1.01)
plt.tight_layout()
plt.savefig('qvalue_sweep_1d.pdf', bbox_inches='tight')
plt.show()
print('Saved qvalue_sweep_1d.pdf')
```

### 5.2  2-D Q-value heatmaps — $Q(\text{Stay})$ and $Q(\text{Stay}) - Q(\text{best move})$ over $(f_s, f_o)$

```python
FS_2DQ = np.linspace(0, 1, 13)
FO_2DQ = np.linspace(0, 1, 13)
N_Q2D  = 20

def qval_sweep_2d(net, agent_type, n_samples=N_Q2D):
    """
    Returns two arrays of shape (len(FO_2DQ), len(FS_2DQ)):
      q_stay   : mean Q(Stay)
      advantage: mean [Q(Stay) - max Q(move)]
    """
    rng = np.random.default_rng(13)
    q_stay   = np.full((len(FO_2DQ), len(FS_2DQ)), np.nan)
    adv      = np.full_like(q_stay, np.nan)
    for j, fs in enumerate(FS_2DQ):
        for i, fo in enumerate(FO_2DQ):
            if fs + fo > 1.0 + 1e-6:
                continue
            qs_acc = np.zeros(N_ACTIONS)
            for _ in range(n_samples):
                p       = make_patch(agent_type, fs, fo, rng=rng)
                qs_acc += get_qvals(net, p)
            qs_mean    = qs_acc / n_samples
            q_stay[i, j] = qs_mean[4]
            adv[i, j]    = qs_mean[4] - np.max(qs_mean[:4])
    return q_stay, adv

print('Computing 2-D Q-value sweeps (this may take ~45 s)...')
qstay_A, adv_A = qval_sweep_2d(net_A, 'A')
qstay_B, adv_B = qval_sweep_2d(net_B, 'B')
print('Done.')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

extent = [FS_2DQ[0], FS_2DQ[-1], FO_2DQ[0], FO_2DQ[-1]]
fs_b   = np.linspace(0, 1, 100)

panels = [
    (axes[0,0], qstay_A, 'RdYlGn', 'Type A — Q(Stay)',           None),
    (axes[0,1], adv_A,   'RdYlGn', 'Type A — Stay advantage',    0.0),
    (axes[1,0], qstay_B, 'RdYlGn', 'Type B — Q(Stay)',           None),
    (axes[1,1], adv_B,   'RdYlGn', 'Type B — Stay advantage',    0.0),
]

for ax, data, cmap, title, center in panels:
    masked = np.ma.masked_invalid(data)
    vmax   = float(np.nanmax(np.abs(data))) if center is not None else float(np.nanmax(data))
    vmin   = -vmax if center is not None else float(np.nanmin(data))

    im = ax.imshow(masked, origin='lower', aspect='auto',
                   extent=extent, cmap=cmap, vmin=vmin, vmax=vmax,
                   interpolation='bilinear')
    plt.colorbar(im, ax=ax, shrink=0.85,
                 label='Q(Stay)' if center is None else 'Q(Stay) − Q(best move)')

    if center is not None:
        # Contour at 0 (indifference between stay and move)
        ax.contour(np.linspace(FS_2DQ[0], FS_2DQ[-1], masked.shape[1]),
                   np.linspace(FO_2DQ[0], FO_2DQ[-1], masked.shape[0]),
                   masked, levels=[0], colors='black', linewidths=2)

    # Schelling boundary
    ax.axvline(SCHELLING_THETA, color='black', lw=2, ls='--', label=f'Schelling θ={SCHELLING_THETA}')
    # Feasibility boundary
    ax.plot(fs_b, 1 - fs_b, 'k:', lw=1.2)
    ax.fill_between(fs_b, 1 - fs_b, 1.0, alpha=0.12, color='gray')

    ax.set_xlabel('Same-type fraction $f_s$', fontsize=10)
    ax.set_ylabel('Opposite-type fraction $f_o$', fontsize=10)
    ax.set_title(title, fontsize=11)
    ax.set_xlim(0, 1); ax.set_ylim(0, 1)
    ax.legend(fontsize=7.5, loc='upper right')

plt.suptitle('2-D Q-value Landscape over $(f_s, f_o)$ space', fontsize=14, y=1.01)
plt.tight_layout()
plt.savefig('qvalue_sweep_2d.pdf', bbox_inches='tight')
plt.show()
print('Saved qvalue_sweep_2d.pdf')
```

# Analysis 3 — Movement direction preference when choosing to move

Given that the agent decides to move ($f_s$ is low), which direction does the DQN prefer? We compute the **softmax distribution** over the four movement actions (excluding Stay) as a function of $f_s$, to check whether the DQN learns a directional preference or treats all directions equivalently (as the Schelling model implicitly assumes).

```python
def move_dir_sweep(net, agent_type, f_opp=0.0, n_samples=N_Q):
    """
    For each f_s, compute softmax over the 4 movement Q-values.
    Returns array (len(FS_FINE), 4).
    """
    rng = np.random.default_rng(17)
    out = np.zeros((len(FS_FINE), 4))
    for k, fs in enumerate(FS_FINE):
        fo = min(f_opp, max(0.0, 1 - fs - 1/N_NEIGH))
        qs_move = np.zeros(4)
        for _ in range(n_samples):
            p = make_patch(agent_type, fs, fo, rng=rng)
            q = get_qvals(net, p)
            qs_move += q[:4]
        qs_mean = qs_move / n_samples
        # softmax
        e = np.exp(qs_mean - np.max(qs_mean))
        out[k] = e / e.sum()
    return out

print('Computing movement direction preference sweeps...')
dir_A = move_dir_sweep(net_A, 'A')
dir_B = move_dir_sweep(net_B, 'B')
print('Done.')
```

```python
DIR_COLORS = ['#1f77b4','#ff7f0e','#2ca02c','#9467bd']

fig, axes = plt.subplots(1, 2, figsize=(13, 4.5), sharey=True)

for ax, dirs, col, label in [
    (axes[0], dir_A, COL_A, 'Type A'),
    (axes[1], dir_B, COL_B, 'Type B'),
]:
    bottom = np.zeros(len(FS_FINE))
    for a in range(4):
        ax.fill_between(FS_FINE, bottom, bottom + dirs[:, a],
                        color=DIR_COLORS[a], alpha=0.8, label=ACTION_NAMES[a])
        bottom += dirs[:, a]

    ax.axvline(SCHELLING_THETA, color='black', lw=2, ls='--',
               label=f'Schelling θ={SCHELLING_THETA}')
    ax.axhline(0.25, color='white', lw=0.8, ls=':')

    ax.set_xlabel('$f_s$ (same-type fraction)', fontsize=10)
    ax.set_ylabel('Softmax share of movement Q-values', fontsize=10)
    ax.set_title(f'{label} — Direction preference\n(conditioned on moving)', fontsize=11)
    ax.set_xlim(0, 1); ax.set_ylim(0, 1)
    ax.legend(fontsize=8, loc='upper right', ncol=2)
    ax.grid(alpha=0.2)

plt.suptitle('Movement Direction Preference vs. Neighbourhood Composition', fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig('move_direction_preference.pdf', bbox_inches='tight')
plt.show()
print('Saved move_direction_preference.pdf')
```

# Analysis 4 — Quantitative comparison table

We summarise:
- Estimated DQN **switch threshold** $\hat{\theta}_{DQN}$ (the $f_s$ at which Q(Stay) first exceeds the best movement Q)
- **Agreement rate** with Schelling rule at $\theta=1/3$
- Whether the DQN threshold depends on $f_o$ (Schelling rule does not)

```python
def estimate_switch_threshold(qvals_1d):
    """Find f_s where Stay advantage first turns positive."""
    adv = qvals_1d[:, 4] - np.max(qvals_1d[:, :4], axis=1)
    crossings = np.where(np.diff(np.sign(adv)))[0]
    if len(crossings) == 0:
        return None
    # Interpolate
    k = crossings[0]
    if adv[k+1] == adv[k]:
        return FS_FINE[k]
    t = -adv[k] / (adv[k+1] - adv[k])
    return FS_FINE[k] + t * (FS_FINE[k+1] - FS_FINE[k])

theta_A = estimate_switch_threshold(qA_1d)
theta_B = estimate_switch_threshold(qB_1d)

print('=' * 60)
print(f'  Schelling threshold (reference)     θ = {SCHELLING_THETA:.3f}')
print('=' * 60)
print(f'  Type A DQN switch threshold  θ̂_A  = {theta_A:.3f}' if theta_A else '  Type A: no clean crossing found')
print(f'  Type B DQN switch threshold  θ̂_B  = {theta_B:.3f}' if theta_B else '  Type B: no clean crossing found')
print('-' * 60)
print(f'  Agreement A vs. Schelling rule      = {agree_A:.1%}')
print(f'  Agreement B vs. Schelling rule      = {agree_B:.1%}')
print('=' * 60)

if theta_A and theta_B:
    diff_A = theta_A - SCHELLING_THETA
    diff_B = theta_B - SCHELLING_THETA
    print(f'\n  Δθ_A = {diff_A:+.3f}  ({"more tolerant" if diff_A < 0 else "more demanding"} than Schelling)')
    print(f'  Δθ_B = {diff_B:+.3f}  ({"more tolerant" if diff_B < 0 else "more demanding"} than Schelling)')
```

# Interpretation guide

| Observation | What it means |
|---|---|
| DQN switch threshold ≈ Schelling θ | The DQN has **rediscovered** the analytic Schelling rule purely from reward signal |
| DQN switch threshold < Schelling θ | DQN agents are **more tolerant** of diverse neighbourhoods (lower homophily requirement) |
| DQN switch threshold > Schelling θ | DQN agents are **more demanding** (require more same-type neighbours before staying) |
| Q(Stay) – Q(move) smooth vs. sharp | Smooth crossing ⟹ soft threshold / gradient-following policy; sharp ⟹ close to Schelling step |
| 2-D map: P(Stay) depends on $f_o$ | DQN exploits information **ignored by Schelling** — opposite-type density matters |
| Direction preference uniform | DQN has not learned spatial biases; movement direction is noise |
| Direction preference structured | DQN has learned **directional escape** strategies (e.g., move away from opposite cluster) |
